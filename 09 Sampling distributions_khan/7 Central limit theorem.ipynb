```python
import numpy as np
import matplotlib.pyplot as plt

population = np.array([1,1,1,1,2,2,3,4,4,5,5,5,5,5])
mu = population.mean()
#mu = np.sum(population) / population.shape[0]
print(mu)

plt.hist(population, bins=np.arange(0,7))
plt.show()
```

```python
import random

population = list(population)

samples_of_x_bar = []

k = 1
for _ in range(1000):
    x_bar = np.array(random.sample(population, k=k)).mean()
    samples_of_x_bar.append(x_bar)
    
plt.hist(samples_of_x_bar)
plt.show()
```

```python
import numpy as np
import scipy.stats as stats

n = 64
mu = 250
sigma = 8 

mu_x_bar = mu
sigma_x_bar = sigma / np.sqrt(n)

left_tail = stats.norm.cdf(mu_x_bar-1.0, loc=mu_x_bar, scale=sigma_x_bar)
right_tail = left_tail
prob_between = 1 - left_tail - right_tail 
print(prob_between)
```
