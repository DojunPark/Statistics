```python
import numpy as np
import matplotlib.pyplot as plt

n = 6
p = 0.2
n_sim = 10000
x = np.random.binomial(1,p,size=(n_sim,n))
x = np.sum(x, axis=1)

bins = np.array([0,1,2,3,4,5,6,7]) - 0.5
plt.hist(x, bins=bins, density=True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

n = 6
p = 0.2

for i in range(20):
    n_sum = i+1
    n_sim = 10000
    x = np.random.binomial(n,p,size=(n_sim,n_sum))
    x = np.sum(x, axis=1)
    mu = n*p*n_sum
    sigma = np.sqrt(n*p*(1-p)*n_sum)
    bins = np.linspace(mu-3*sigma,mu+3*sigma,29)
    plt.hist(x, bins=bins, density=True)
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

for i in range(20):
    n_sum = i+1
    n_sim = 10000
    x = np.random.uniform(0,1,size=(n_sim,n_sum))
    x = np.sum(x, axis=1)
    mu = (1/2)*n_sum
    sigma = np.sqrt((1/12)*n_sum)
    bins = np.linspace(mu-3*sigma,mu+3*sigma,29)
    plt.hist(x, bins=bins, density=True)
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

for i in range(20):
    n_sum = i+1
    n_sim = 10000
    x = np.random.exponential(size=(n_sim,n_sum))
    x = np.sum(x, axis=1)
    mu = n_sum
    sigma = np.sqrt(n_sum)
    bins = np.linspace(mu-3*sigma,mu+3*sigma,29)
    plt.hist(x, bins=bins, density=True)
    plt.show()
```
